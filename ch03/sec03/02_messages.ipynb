### 메시지 (Messages)

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage, AIMessage, SystemMessage

model = init_chat_model("google_genai:gemini-2.5-flash-lite")

system_msg = SystemMessage("You are a helpful assistant.")
human_msg = HumanMessage("Hello, how are you?")

# Use with chat models
messages = [system_msg, human_msg]
response = model.invoke(messages)  # Returns AIMessage
```

```python
response
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    # 시스템 메시지: 모델에게 '시(詩) 전문가' 역할을 부여합니다.
    SystemMessage("당신은 시 전문가입니다"),
    
    # 인간 메시지: 사용자 입력으로, '봄에 관한 하이쿠를 작성해달라'는 요청입니다.
    HumanMessage("봄에 관한 하이쿠를 작성해줘"),
    
    # AI 메시지: 모델이 이전에 응답했던 내용으로 대화 기록에 추가됩니다.
    AIMessage("벚꽃이 피어나네...")
]
# 이 메시지 목록을 모델에 전달하여 다음 응답을 얻습니다.
response = model.invoke(messages)
```

```python
response
```

```python
# Dictionary format
messages = [
    {"role": "system", "content": "당신은 시 전문가입니다"},
    {"role": "user", "content": "봄에 관한 하이쿠를 작성해줘"},
    {"role": "assistant", "content": "벚꽃이 피어나네..."}
]
response = model.invoke(messages)
```

```python
response
```

```python
from langchain.messages import SystemMessage, HumanMessage

# 비교의 편의상 답변 길이 제한을 추가합니다.
system_msg = SystemMessage("당신은 도움이 되는 코딩 보조 전문가입니다. 500자 이내로 답변하세요.")

messages = [
    system_msg, 
    HumanMessage("REST API를 어떻게 만드나요?")
]

# 이 메시지 목록을 모델에 전달하여 응답을 얻습니다.
response = model.invoke(messages)
```

```python
response
```

```python
print(response.content)
```

```python
from langchain.messages import SystemMessage, HumanMessage

# 시스템 메시지: 모델의 역할, 응답 스타일, 지침을 상세하게 설정합니다.
system_msg = SystemMessage("""
당신은 웹 프레임워크에 전문성을 가진 시니어 파이썬 개발자입니다.
응답은 반드시 Flask를 사용한 REST API의 핵심 CRUD 구현 방법 5가지를 요약한 표로 시작하십시오.
이 표 다음에 전체 코드 예시와 당신의 추론 과정을 상세히 설명하십시오.
설명은 간결하지만 철저하게 하십시오.
500자 이내로 답변하세요.
""")

# 전체 메시지 목록: 시스템 지침과 사용자 질문을 포함합니다.
messages = [
    system_msg, # 모델의 상세 역할과 행동 방식을 지시하는 시스템 메시지
    HumanMessage("REST API를 어떻게 만드나요?") # 사용자가 모델에 하는 질문
]

# 이 메시지 목록을 모델에 전달하여 응답을 얻습니다.
response = model.invoke(messages)
```

```python
print(response.content)
```
