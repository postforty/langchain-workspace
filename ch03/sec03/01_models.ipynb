```python
from langchain.chat_models import init_chat_model
import os
gemini_api_key = os.getenv("GEMINI_API_KEY")

model = init_chat_model(
    model="google_genai:gemini-2.5-flash-lite", 
    api_key=gemini_api_key,
    temperature=0.7,
    timeout=30,
    max_tokens=1000
)
```

```python
response = model.invoke("앵무새들은 왜 말하는 거야?") # 문자열 전달
```

```python
response
```

```python
conversation = [
    {"role": "system", "content": "당신은 영어를 한글로 번역하는 친절한 조수입니다."},
    {"role": "user", "content": "I love programming."},
    {"role": "assistant", "content": "프로그래밍을 사랑합니다."},
    {"role": "user", "content": "I love building applications."}
]

response = model.invoke(conversation)
print(response)
```

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

conversation = [
    SystemMessage("당신은 영어를 한글로 번역하는 친절한 조수입니다."),
    HumanMessage("I love programming."),
    AIMessage("프로그래밍을 사랑합니다."),
    HumanMessage("I love building applications.")
]

response = model.invoke(conversation)
print(response)
```

```python
next(model.stream("앵무새는 왜 깃털이 화려할까요?"))
```

```python
for chunk in model.stream("앵무새는 왜 깃털이 화려할까요?"):
    print(chunk.text, end="|", flush=True)
```

```python
responses = model.batch([
    "앵무새는 왜 깃털이 화려한가요?",
    "비행기는 어떻게 하늘을 날까요?",
    "양자 컴퓨팅이란 무엇인가요?"
])
for response in responses:
    print(response)
```
