```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0, # 창의성 0 ~ 2.0
    google_api_key=gemini_api_key
)
```

```python
prompt = """당신은 LangChain 개발자입니다. 새로운 에이전트(Agent)를 구축하는 과정을 담은 매우 짧은 기술 동화 또는 비유적인 이야기를 작성하시오.
1. 이야기는 반드시 120자 이상 250자 이하의 한글로 구성되어야 합니다.
2. 이야기에 'Chain', 'PromptTemplate', 'Memory' 세 단어를 반드시 포함해야 합니다.
3. 이야기의 결말은 구축한 에이전트가 현실 세계에서 **예상치 못한** 문제 해결에 성공하는 장면으로 마무리되어야 합니다."""
```

```python
response = llm.invoke(prompt)
```

```python
response.content
```
